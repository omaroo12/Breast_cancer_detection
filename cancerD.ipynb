```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data.csv')
df
```

```python
df.info()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.drop(['Unnamed: 32','id'],axis=1,inplace=True)
df
```

```python
df['diagnosis'].value_counts()
```

```python
sns.countplot(data=df, x='diagnosis')
plt.show()
```

```python
typee=pd.get_dummies(df['diagnosis'])
typee
```

```python
df=pd.concat([df,typee],axis=1)
df
```

```python
df.drop(['M','diagnosis'],axis=1,inplace=True)
```

```python
df.describe().T
```

```python
plt.figure(figsize=(20,20))
sns.heatmap(df.corr(),annot=True)
```

```python
features=df.drop('B',axis=1)
target=df['B']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
```

```python
X_train,X_test,y_train,y_test=train_test_split(features,target,test_size=.2,random_state=42)
```

```python
s=StandardScaler()
X_train=s.fit_transform(X_train)
X_test=s.fit_transform(X_test)
```

```python
model=LogisticRegression()
model.fit(X_train,y_train)
pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score


```

```python
sc=accuracy_score(y_test,pred)
sc
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train,y_train)
pred2=knn.predict(X_test)
```

```python
sc2=accuracy_score(y_test,pred2)
sc2
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
rf.fit(X_train,y_train)
pred3=rf.predict(X_test)
```

```python
sc3=accuracy_score(y_test,pred3)
sc3
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(random_state=42).fit(X_train,y_train)
pred4=dt.predict(X_test)
```

```python
sc4=accuracy_score(y_test,pred4)
sc4
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gn=GaussianNB()
gn.fit(X_train,y_train)
pred5=gn.predict(X_test)
```

```python
sc5=accuracy_score(y_test,pred5)
```
